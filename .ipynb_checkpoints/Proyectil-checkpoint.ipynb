### Modelo de vuelo y impacto
Cómo primer modelo de estudio se considera el lanzamiento de un proyectil, hacia un onjetivo perfectamente cuadrado de lados \emp{l} a una distancia horizontal $d$ del punto de lanzamiento. Éste proyectil se mueve en un medio (aire) que presenta fricción viscosa, donde ésta fuerza viene dada por la expresión:
\begin{equation}
\vec{F}_{fv}= -\lambda \vec{v} - \beta v \vec{v}
\label{viscosa}
\end{equation} 
Donde $\lambda$ suministra la información de la fricción no turbulenta del aire, y $\beta$ suministra la información turbulenta.

```python
parametros={'m':1.0,'lam':0.9,'bti':0,'btf':0.05,'vi':12,'theta': np.pi/4,'h':1e-2,'d':5.5,'l':1.0}
```

```python
class Particula: 
    """Ésta clase contiene la información sobre la trayectoria de la particula, posiciones, velocidades
       parametros de lanzamiento y si impacta el objetivo, posee un metodo que determina la informacion
       estadistica de la trayectoria, y las graficas de trayectoria, zona de evalucion y energias."""
    def __init__(self,x,y,v_x,v_y,t,impact,prm):
        self.posx=x    ; self.posy=y
        self.velx=v_x  ; self.vely=v_y
        self.tiempo=t  ; self.impact= impact ; self.prm=prm
        
    def estadisticas(self):
        self.xmax= self.posx.max()                   ;   self.hmax= self.posy.max()
        self.timup= self.tiempo[self.posy.argmax()]  ;   self.timfly= self.tiempo.max()
        
    def Grafica(self):
        plt.plot(self.posx,self.posy) ; plt.xlabel("Desplazamiento en $x$",fontsize=15) 
        plt.ylabel("Desplazamiento en $y$",fontsize=15) ; plt.title("Trayectoria de tiro",fontsize=15) 
        plt.tick_params(axis="both",labelsize=12) ; plt.grid(True)
        return plt.show()
    
    def Target_zone(self):
        plt.plot(self.posx,self.posy,'.')                 ; prm=self.prm
        #Objetivo 
        plt.xlim(prm['d']-2*prm['l'],prm['d']+2*prm['l']) ; plt.ylim(0,2*prm['l'])
        p1x=prm['d']                                      ; p1y=0.0 ; p2x=prm['d']+ prm['l']   
        p2y=p1y ; p3x=p2x ; p3y=prm['l'] ; p4x=p1x  ; p4y=p3y 
        x=[p1x,p2x,p3x,p4x,p1x] ; y=[p1y,p2y,p3y,p4y,p1y]
        plt.plot(x,y)
        #Zona de evalución
        z1x=prm['d']-0.05 ; z1y=0.0 ; z2x=prm['d']+ prm['l'] + 0.05 ; z2y=z1y 
        z3x=z2x ; z3y=prm['l'] + 0.025 ; z4x=z1x ; z4y=z3y 
        x1=[z1x,z2x,z3x,z4x,z1x] ; y1=[z1y,z2y,z3y,z4y,z1y]
        plt.plot(x1,y1,'--') ;plt.xlabel("$x$") ; plt.ylabel("$y$")
        plt.title("Zona del objetivo");plt.grid(True);plt.legend(("Trayectoria","objetivo","Zona de evalución"))
        return plt.show(),self.impact
    
    def Energia(self):
        k=(self.prm["m"]/2)*(self.velx**2+self.vely**2) ; V=self.prm["m"]*(9.8)*self.posy ; E=k+V
        plt.plot(self.tiempo,k,"blue",self.tiempo,V,"red",self.tiempo,E,"--")
        plt.xlabel("Tiempo") ; plt.ylabel("Energia") 
        plt.title("Energias vs tiempo");plt.grid(True);plt.legend(("Cinetica","Pontencial","Energia Total"))

def Launch_projectile(prm):
    """Ésta función realiza un lanzamiento del proyectil (evolucion de sistema), utilizando el runge kutta 
       de cuarto orden,teniendo en cuenta los parametros de entrada, y regresa una instancia de la clase 
       particula."""
    
    #condiciones iniciales 
    ro=np.array([0,0])      ; vo=np.array([prm['vi']*np.cos(prm['theta']),prm['vi']*np.sin(prm['theta'])])
    yo=np.hstack((ro,vo))
    
    #Función vectorial de derivadas (dx/dt, dv/dt)
    """Se define una funcion que representa la derivada del vector dinamico, la cual sera integrada por el 
       runge kutta"""
    def dydt_p(y,beta,m=prm['m'],lam=prm['lam']/prm['m']):
        g=-9.8   ;   vv=np.sqrt((y[2:4]**2).sum())
        return np.array([ y[2], y[3] , -(1/m)*(lam*y[2]+beta*vv*y[2]), g-(1/m)*(lam*y[3]+beta*vv*y[3]) ])
    
    def rk4_p(f,yo,h=prm['h'],bti=prm['bti']/prm['m'],btf=prm['btf']/prm['m']): #yo=(x_o,y_o,vx_o,vy_o)
        """Esta funcion realiza la evolucion del sistema utilizando el runge kutta de cuarto orden,
           regresa la derivada del vector dinamico, el arreglo de tiempos y si hubo impacto"""
        
        #El sistema inicia en tiempo cero y desde las condiciones iniciales
        t=0       ;      x=np.array([yo]) ;      tiempo=[0] ;  impact="None"
        
        """Evoluciona desde t=0 hasta que la componene en Y de posicion sea menor que cero"""
        while x[-1][1]>0 or t==0:
            beta=np.random.uniform(bti,btf)
            k1 = h*f(x[-1],beta)               ;   k2 = h*f(x[-1]+0.5*k1,beta)
            k3 = h*f(x[-1]+0.5*k2,beta)        ;   k4 = h*f(x[-1]+k3,beta)
            x = np.vstack(( x, x[-1]+(k1+2*(k2+k3)+k4)/6 )) 
            t+=h ; tiempo=np.hstack((tiempo,t))
            
        #impacto con el objetivo, si la posicion de la particula esta dentro de las dimensiones del objetivo  
            if x[-1][1] <= prm['l'] and prm['d'] <= x[-1][0] <= prm['d'] + prm['l'] and prm['l'] != 0:
                impact="Target" ;break  
                
        def recta(v1,v2,y):
            """Ésta funcion realiza la interpolacion entre dos puntos y encuentra la componente en x 
               de la recta que pasa por un "y" dado"""
            m=(v2[1]-v1[1])/(v2[0]-v1[0]) ; b=v2[1]-m*v2[0]
            return np.array([(y-b)/m,y])
        
        """En esta parte se toman las ultimas 5 posiciones de la particula y se verifica si estan dentro de la
           region de valuacion, si es asi, se realiza una interpolacion entre parejas de puntos, con el 
           objetivo de verificar si la particula impacto al objetivo"""
        if impact!="Target" and prm['l'] != 0:
            ve=x[-5:] ; xx=ve[:,0] ; yy=ve[:,1]   
            for i in range(1,len(ve)):
                if (prm['d']-0.05 <= xx[i-1] <= prm['d'] + prm['l'] + 0.05 
                    and  prm['d']-0.05 <= xx[i] <= prm['d'] + prm['l'] + 0.05 
                    and yy[i-1] <= prm['l']+0.025 and yy[i] <= prm['l']+ 0.025):
                    v1=[xx[i-1],yy[i-1]] ;  v2=[xx[i],yy[i]]
                    Point=recta(v1,v2,prm['l'])
                    if prm['d'] <= Point[0] <= prm['d'] + prm['l']:
                        impact="Target"
                        x[-1][0:2]=Point
                        for ii in range(i,len(ve)):
                            x[-ii][0:2]=Point
                        break    
        #impacto con el suelo. 
        """si la posicion en y es menor que cero, se realiza una interpolacion para encontrar el punto
           donde corta a y=0, es decir donde impacto con el suelo"""      
        if x[-1][1] < 0:
            v1=x[-2][0:2] ; v2=x[-1][0:2] ; x[-1][0:2]=recta(v1,v2,0.0); impact="Floor"
            
            
        return x,tiempo,impact 
    #Se realiza la evolucion del sistema
    Y=rk4_p(dydt_p,yo)
    """Regresa una instancia de la clase particula, con los datos calculados"""
    return Particula(Y[0][:,0],Y[0][:,1],Y[0][:,2],Y[0][:,3],Y[1],Y[2],prm)
```

```python
parametros={'m':1.0,'lam':0.9,'bti':0,'btf':0.05,'vi':12,'theta': 45*(np.pi/4)/45,'h':1e-2,'d':6.1,'l':0.1}
Proyectil=Launch_projectile(parametros)
```

### Un sistema no determinista
A continuación se muestra la manera como se realizan N tiros son un set de parametros constante.

```python
def Experimento_proyectil(N,parametros):
    """Ésta funcion realiza N lanzamientos con una set de parametros constantes, entrega la informacion 
       estadistica de cada tiro, y la probabilidad de impacto estimada."""
    proyc=Launch_projectile(parametros)
    proyc.estadisticas()
    datos=np.array([proyc.hmax,proyc.xmax,proyc.timfly,proyc.timup])
    shoot=0
    if proyc.impact=="Target": shoot+=1
    for i in range(1,N):
        proyc=Launch_projectile(parametros)
        proyc.estadisticas()
        datos= np.vstack((datos,np.array([proyc.hmax,proyc.xmax,proyc.timfly,proyc.timup])))
        if proyc.impact=="Target": shoot+=1
    prob=(shoot/N)*100      
    return datos,prob
```

```python
parametros={'m':1.0,'lam':0.9,'bti':0,'btf':0.05,'vi':12,'theta': 31*(np.pi/4)/45,'h':1e-2,'d':6.3,'l':0.1}
Experimento_proyectil(1000,parametros)
```

```python
def Data_experiment(N,parametros):
    """Ésta funcion realiza los N lanzamientos con un set de patametros constante al llamar a la funcion 
       "Experimento_proyectil",toma la infomacion estadistica de cada los N tiros y realiza, los promedios
       y desviaciones estadar de ellos. (hmax,xmax,time-fly,time-up). Regresa los promedios y desviaciones 
       y el arreglo de los alcances maximos para los N lanzamientos"""
    Exp=Experimento_proyectil(N,parametros)[0]
    hmax,hmax_sig= Exp[:,0].mean(), Exp[:,0].std()
    xmax, xmax_sig= Exp[:,1].mean(), Exp[:,1].std()
    time_fly, time_fly_sig= Exp[:,2].mean(), Exp[:,2].std()
    timeup, timeup_sig= Exp[:,3].mean(), Exp[:,3].std()
    data=np.array([[hmax,hmax_sig],[xmax, xmax_sig],[time_fly, time_fly_sig],[timeup, timeup_sig] ])
    return data,Exp[:,1]
```

```python
parametros={'m':1,'lam':0.9,'bti': 0,'btf':0.05,'vi':12,'theta': 34*(np.pi/4)/45,'h':1e-2,'d':6.0,'l':0}
hh=Data_experiment(1000,parametros)
hh[0]
```

### Distribucion de alcance maximo

```python
def distribucion(N,parametros,intervalos):
    """Ésta funcion toma el arreglo de los alcances maximos para los N lanzamientos, realizados por la 
       funcion "Data_experiment", y realiza el histograma de los alcances maximos. Tambien toma los puntos
       medios de los intervalos del histograma en conjuto con la altura de los bins, y realiza un fitting."""
    Data=Data_experiment(N,parametros)[1]
    bins_y,bins_x=np.histogram(Data,intervalos)
    p_medios_intervalos=np.array([ (bins_x[i-1]+bins_x[i])/2  for i in range(1,len(bins_x)) ])
    
    def Gaus(x,mean,sigma,b): return (1/b)*np.e**(-(x-mean)**2 /(2*sigma**2))
    fit=op.curve_fit(Gaus,p_medios_intervalos,bins_y)
    mean,sigma,b=fit[0]
    
    plt.hist(Data,intervalos)
    plt.plot(p_medios_intervalos,Gaus(p_medios_intervalos,mean,sigma,b)) 
    plt.xlabel("Alcance maximo",fontsize=15) ; plt.ylabel("Numero de tiros",fontsize=15) 
    plt.title("Distribución del alcance",fontsize=15); plt.tick_params(axis="both",labelsize=12)
    plt.show()
    return "mean:",mean,"sigma:",sigma
```

```python
parametros={'m':1,'lam':0.9,'bti': 0,'btf':0.05,'vi':12,'theta': 45*(np.pi/4)/45,'h':1e-2,'d':6.0,'l':0}
distribucion(10000,parametros,45) 
```

### Variacion del alcance maximo en funcion del angulo de tiro

```python
def range_funcion_theta(theta_i,theta_f,delta_theta,prm,N):
    """Ésta funcion realiza N lanzamientos por angulo, barriendo desde theta-inical hasta theta-final,
       almacena el alcance maximo promedio de cada tiro N tiros por angulo, y regresa la grafica de 
       alcance maximo en funcion del angulo de tiro, tambien regresa el angulo con mayor alcance y su alcance"""
    range_theta=np.linspace(theta_i,theta_f,91) ; range_x=[] ; std=[]
    for i in range(len(range_theta)):
        prm['theta']=range_theta[i]
        data= Data_experiment(N,prm)
        range_x= np.hstack((range_x,data[1][0]))
        std=np.hstack((std,data[1][1]))
    #grafica    
    plt.plot(range_theta,range_x); plt.xlabel("Angulo de tiro [rad]",fontsize=15)
    plt.ylabel("Alcance maximo",fontsize=15) ; plt.title("Alcance maximo vs Angulo de tiro",fontsize=15)
    plt.tick_params(axis="both",labelsize=12); plt.grid(True)
    better_theta= range_theta[range_x.argmax()]
    
    return np.array([better_theta*45/(np.pi/4),range_x.max()])
```

```python
parametros={'m':1,'lam':0.9,'bti': 0,'btf':0.01,'vi':12,'theta': 45*(np.pi/4)/45,'h':1e-2,'d':6.0,'l':0}
range_funcion_theta(0,np.pi/2,1,parametros,1000)
```

### Angulo optimo de tiro
Se barren angulos 15 a 60 grados, y se encuentra el angulo que maximiza la probabilidad de impacto con el objetivo a una distancia fija d. 

```python
def Impact(size,d,prm,N):
    """Ésta funcion realiza un barrido en el angulo de tiro de 15 hasta 60 grados, realizando por angulo
       N lanzamientos. Regresa la probabilidad de cada impacto para angulo con el onjetivo a una distancia fija,
       tambien regresa el angulo con mayor probabilidad, y la porbabilidad de impacto de éste."""
    prm['d']= d ;prm['l']=size ; theta=np.linspace(np.pi/12,np.pi/3,46); prob=[]
    for i in range(len(theta)):
        prm['theta']=theta[i]
        prob=np.hstack((prob,Experimento_proyectil(N,prm)[1]))
    return prob,prob.max(),theta[prob.argmax()]*45/(np.pi/4)    
```

```python
parametros={'m':1.0,'lam':0.9,'bti':0,'btf':0.05,'vi':12,'theta': 45*(np.pi/4)/45,'h':1e-2,'d':6.5,'l':0.1}
Impact(0.1,5.8,parametros,1000)
```

### Pareja de angulo y distancia optima de tiro
Se barren distancias en un intervalo, y por cada una se barren angulos de 15 y 60 grados, realizando N tiros por angulo. 

```python
def Run(size,prm,N,xi,xf,delta): #PRECAUCION-el codigo tiene un tiempo estimado de 2 horas 15 minutos. 
    """Ésta funcion realiza un barrido de la distancia al objetivo desde xi hasta xf, realizando por cada
       distancia un llamada a la funcion Impact, que realiza un barrido en el angulo de tiro de 15 hasta 60 
       grados, realizando por angulo N lanzamientos. Regresa una matriz de 3 columnas, donde se tiene, 
       la distancia a la que se hizo el lanzamiento, la probabilidad de impacto mayor y el angulo que 
       la genera, con una set de parametros dada."""
    re=np.array([0,0,0])
    while xi <= xf:
        k=Impact(size,xi,prm,N)
        re=np.vstack((re,np.array([xi,k[2],k[1]])))
        xi+=delta
    return re    
```

```python
parametros={'m':1.0,'lam':0.9,'bti':0,'btf':0.05,'vi':12,'theta': 45*(np.pi/4)/45,'h':1e-2,'d':6.5,'l':0.1}
Run(0.1,parametros,1000,5.0,6.5,0.1)
```
